        <p>Primeiramente, agradecemos pelo seu interesse em fazer parte do time de Data Science da Beluga!</p>

<p>Preparamos um <em>coding challenge</em> para testar algumas habilidades básicas importantes para avaliar a sua capacidade de enfrentar os desafios que temos no dia-a-dia. Abaixo descrevemos o contexto do problema e o que esperamos de você ao final do <em>challenge</em>.</p>

<p>Uma grande empresa de telecomunicações possui diversos equipamentos de mensuração de sinais em campo espalhado pelo Brasil. Atualmente eles fazem mensuração remota de logs de erro dos equipamentos e baseado nos mesmos equipes técnicas fazem sua análise. Baseado em sua experiência as equipes julgam necessidade de visita técnica para realizar reparos/manutenções nos equipamentos.</p>

<p>O gestor da área de manutenções tem notado que muitas das visitas técnicas realizadas por sua equipe não eram de fato necessárias e está buscando meios de minimizar esses casos.</p>

<p>Nesse contexto o gestor procurou a Beluga para ajudá-lo a endereçar esse problema.</p>

<p>Para tal ele nos forneceu os dados passados de alarmes (<code class="highlighter-rouge">training dataset</code>) com diversas informações referentes ao equipamento (<code class="highlighter-rouge">features</code>) e com a indicação se era ou não necessária a visita técnica (<code class="highlighter-rouge">target</code>).</p>

<p>Baseado nesses dados se espera que você como Data Scientist Beluga consiga gerar um modelo para prever necessidade de visita técnica em alarmes futuros (<code class="highlighter-rouge">test dataset</code>).</p>

<p>Você pode nos entregar o resultado do seu trabalho de várias formas! Abaixo citamos algumas:</p>

<ul>
  <li>
    <p>Um arquivo csv com as previsões no formato <code class="highlighter-rouge">id</code>,<code class="highlighter-rouge">target</code> (este é obrigatório!)</p>
  </li>
  <li>
    <p>Código fonte da implementação do algoritmo (preferivelmente em um repositório público ou privado no <em>github</em>)</p>
  </li>
  <li>
    <p>Documentação sucinta do projeto</p>
  </li>
  <li>
    <p>Outros documentos que possam mostrar evidências de como você resolveu o problema (enquanto melhor nós consigamos entender o seu processo de trabalho, melhor podemos avaliar o <em>challenge</em>)</p>
  </li>
</ul>

<p>A seguir estão escritos alguns detalhes importantes a respeito dos dados com os quais você irá trabalhar. Lembre-se: dado as limitações de tempo, se você tiver que escolher entre um modelo de precisão perfeita ou um modelo simples mas com dados analisados e tratados de forma certa, prefira segunda opção.</p>

<p>Informações técnicas do conjunto de dados (dataset) do <em>challenge</em>:</p>

<ul>
  <li>
    <p>Os dados são artificias e anonimizados e seus valores não possuem nenhum significado especial.</p>
  </li>
  <li>O <code class="highlighter-rouge">dataset</code> está separado em <code class="highlighter-rouge">training set</code> e um <code class="highlighter-rouge">test set</code>:
    <ul>
      <li>O dataset possui um total de 13 variáveis.</li>
      <li>O <code class="highlighter-rouge">training set</code> possui uma variável <code class="highlighter-rouge">target</code>, que pode ter valor 1 ou 0.</li>
    </ul>
  </li>
  <li>
    <p>Queremos que você manipule os dados, encontre padrões, e no fim crie um modelo para prever o valor da coluna <code class="highlighter-rouge">target</code> do <code class="highlighter-rouge">test set</code>.</p>
  </li>
  <li>
    <p>A métrica de avaliação que será utilizada será:</p>

    <ul>
      <li>Especificidade</li>
    </ul>

```python
def make_hist(training_data,column='feature_1', ymax=None):
    df2 = training_data.groupby(['target', column]).size().unstack('target').fillna(0)
    necessaria_count = df2[1]
    desnecessaria_count = df2[0]
    tot_count = training_data[column].value_counts()

    necessaria_perc = necessaria_count/tot_count
    err_perc = necessaria_count/tot_count*np.sqrt(1/necessaria_count+1/tot_count)
    err_perc[np.isnan(err_perc)] = 1
    x = np.arange(len(tot_count))
    xlabel = df2.index
    x2 = np.hstack([[-1],x,[max(x)+1]])

    plt.figure(figsize=(30,10))
    ax = plt.subplot(111)

    ax.bar(x, necessaria_count, color='g', align='center')
    ax.bar(x, desnecessaria_count, bottom=necessaria_count, color='r', align='center')
    ax.set_ylabel('Visitas', fontsize=20)

    ax2 = ax.twinx()
#     ax2.plot(x2,np.full_like(x2, perc_evad_tot*100), 'b')
#     ax2.errorbar(x, list(necessaria_perc*100), yerr=list(err_perc*100), \
#                  marker='o',mec='b', mfc='b', ls='None', markersize=10)
    ax2.set_ylim([0,100])
    ax2.set_ylabel('% Visitas necessárias', fontsize=20)

    if ymax==None:
        ymax = (max(tot_count)//1000+1)*1000
    ax.set_ylim([0,ymax])
    ax.set_yticks(np.arange(0,ymax+1, ymax/4))

    ax.set_xlim([min(x2), max(x2)])
    ax.set_xticks(x)
    ax.set_xticklabels(xlabel, fontsize=16, rotation=90)
    ax.grid()

    camel_title = column.replace('_', ' ').title()
    ax.set_xlabel(camel_title, fontsize=20)
    plt.title('Visitas Necessárias vs '+camel_title, fontsize=28)
    
    plt.show()
    return ax, ax2
```

```python
features = [c for c in training_data.columns if c not in {'id', 'target', 'feature_0'}]
for col in features:
    print(col, training_data[col].nunique())
    if training_data[col].nunique() < 100:
        make_hist(training_data, col)
```
